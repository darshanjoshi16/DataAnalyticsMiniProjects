# **Data Analytics Mini Project - 1**

Dataset: Udemy Course Dataset

In this notebook, we will analyse the udemy course dataset and will try to give the insights from the dataset by applying the various data analytics methods.

**Section 1: Importing the libraries** 

```python
#importing the libraries to perform the task
import pandas as pd
from google.colab import files
import io
```

**Section 2: Importing the Data**

```python
#using files module we will take csv file as the input
uploading = files.upload()

#creating the dataframe and reading csv and  using IO library for converting into data frame
df = pd.read_csv(io.BytesIO(uploading['Udemy_Courses.csv']))
```

**Section 3: Exploring the Data**

```python
#printing the whole dataset
print(df)
```

**Section 4: Taking the insights using the questions**

**(1). What are the all subjects for which Udemy is offering the courses?**

```python
#it will show different subjects for which courses are being offered
df.subject.unique()
```

**(2). Which subjects has the maximum number of courses?**

```python
#it will search the counts into subject column
df.subject.value_counts()
```

**(3). Show all the courses which are free of cost.**

```python
df[df.is_paid == False]
```

**(4). Show all the courses which are paid.**

```python
df[df.is_paid == True]
```

**(5). Which are the top selling courses?**

```python
#it will sort the tuples as per the number of subscriber in the ascending mode
df.sort_values('num_subscribers',ascending=False)
```

**(6). Which are the least selling courses?**

```python
#it will sort the tuples as per the number of subscriber in the descending mode
df.sort_values('num_subscribers', ascending=True)
```

**(7). Show all the courses of Web development where the price is less than 500**

```python
#using and operator we are doing the comparision
df[(df.subject == 'Web Development') & (df.price < '500')]
```

**(8). List out all the courses which are related to ReactJS.**

```python
#converting the content of course title column into  string and checking the ReactJS is among the string or not
df[df.course_title.str.contains('ReactJS')]
```

**(9). List out the courses published in year 2016**

```python
#converting the timestamp column using the datetime functionality and different methods
df['published_timestamp'] = pd.to_datetime(df.published_timestamp)
df['Year'] = df['published_timestamp'].dt.year
df[df.Year == 2016]
```

**(10). What are the max numbers of subscribers for each level of courses?**

```python
#using groupby method and grouping on the based of level and giving the maximum value
df.groupby('level')['num_subscribers'].max()
```
